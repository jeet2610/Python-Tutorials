#### WebScraping using request module

```python
import requests
import json
response = requests.get('https://jsonplaceholder.typicode.com/posts')

if response.status_code == 200:
    posts = response.json()
    if posts:
        print("UserID \t ID \t Title \t Body\t")
        for i in range(6):
            post = posts[i]
            print(f"{post['userId']} \t {post['id']} \t {post['title']} \t {post['body']} \t \n") 
else:
    print("Failed to Fetch from the server")
```

```python
import requests

response = requests.get('https://jsonplaceholder.typicode.com/comments')
counter = {}
if response.status_code == 200:
    comments = response.json()
    
    for comment in comments:
        comments = comment['postId']
        if comments in counter:
            counter[comments] +=1
        else:
            counter[comments] = 1
    
    for comments, count in counter.items():
        print(f"Post {comments} has {count} comments")
```

```python
from selenium import webdriver
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Navigate to the Wikipedia page
driver.get("https://en.wikipedia.org/wiki/List_of_Indian_IT_companies")

# Wait for the table to load (adjust the time if necessary)
time.sleep(3)  # Using sleep for demonstration, consider using explicit waits

# Find the table by its class name (you might need to adjust this selector based on the page structure)
table = driver.find_element_by_class_name("wikitable")

# Get all rows from the table
rows = table.find_elements_by_tag_name("tr")

# Extracting data from the table rows
for row in rows:
    # Get the data from each cell in the row
    cells = row.find_elements_by_tag_name("td")
    for cell in cells:
        print(cell.text)  # Print or store the text of each cell

# Close the browser
driver.quit()
```
