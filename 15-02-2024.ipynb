```python
import numpy as np

dragon_image = np.zeros((100, 100, 3), dtype=np.uint8)

# Blue flames
blue_flame = np.array([0, 0, 255], dtype=np.uint8)

# Yellow flames
yellow_flame = np.array([255, 255, 0], dtype=np.uint8)

# Draw blue flames
dragon_image[70:100, 40:60] = blue_flame
dragon_image[60:70, 30:70] = blue_flame
dragon_image[50:60, 20:80] = blue_flame

# Draw yellow flames
dragon_image[70:100, 70:90] = yellow_flame
dragon_image[60:70, 60:100] = yellow_flame
dragon_image[50:60, 50:110] = yellow_flame

print(dragon_image)
```

```python
from PIL import Image
img_rgb = Image.open('images/Elon_Musk.jpg')
img_gray = img_rgb.convert('L')
img_gray.save('images/gray.jpg')
```

```python
img_grey = np.array(Image.open('images/gray.jpg'))
img = np.array(Image.open('images/Elon_Musk.jpg'))
fig = plt.figure(figsize=(10,10))
img_grey = 255*3 - img_grey
fig.add_subplot(1,2,1)
plt.imshow(img_grey)
plt.title('Negative of Grey Image')
img = 255-img
fig.add_subplot(1,2,2)
plt.imshow(img)
plt.title('Negative of Rgb img')
```

```python
# Padding Black Spaces
img = np.array(Image.open('images/Elon_Musk.jpg'))
img_grey = img.sum(2)/(255*3)
img0 = img_grey.copy()
img0 = np.pad(img0,((50,50),(50,50)),mode='constant')
plt.imshow(img0)
```

```python
# RGB channels
img = np.array(Image.open('images/Elon_Musk.jpg'))
img_R,img_G,img_B = img.copy(),img.copy(),img.copy()
img_R[:,:,(1,2)]=0
img_G[:,:,(0,2)]=0
img_B[:,:,(1,1)]=0
img_rgb = np.concatenate((img_R,img_G,img_B),axis=1)
plt.figure(figsize=(15,15))
plt.imshow(img_rgb)
```

```python
# Colour Reductions
img_0 = (img//64)*64
img_1 = (img//128)*128
img_all = np.concatenate((img,img_0,img_1),axis=1)
plt.figure(figsize=(15,15))
plt.imshow(img_all)
```
