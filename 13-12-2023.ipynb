#### Normal Distribution
##### LOC - mean
##### Scale - standard Deviation
##### Size - The shape of the returned Array

```python
from numpy import random
x = random.normal(loc=1, scale=2, size=(2, 3))
print(x)
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Mean and standard deviation for heights
mean_height = 175
std_dev = 7

# Generate random data following a normal distribution
heights = np.random.normal(mean_height, std_dev, 1000)

# Create a histogram to visualize the distribution
plt.hist(heights, bins=30, density=True, alpha=0.6, color='g')

# Add a best fit line (normal distribution curve)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_height, std_dev)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Heights')
plt.grid(True)
plt.show()
```

#### Binomial Distribution
##### n - number of trials.
##### p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).
##### size - The shape of the returned array

```python
# Given 10 trials for coin toss generate 10 data points
from numpy import random

x = random.binomial(n=10, p=0.5, size=10)

print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(random.binomial(n=10, p=0.5, size=1000), kde=False)

plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
x = random.normal(loc=50, scale=5, size=1000)
y = random.binomial(n=100, p=0.5, size=1000)
sns.distplot(x, hist=True, label='normal')
sns.distplot(y, hist=True, label='binomial')

plt.show()
```

#### Poisson Distribution
##### It estimates how many times an event can happen in a specified time. e.g. If someone eats twice a day what is the probability he will eat thrice?
###### It has two parameters:

###### lam - rate or known number of occurrences e.g. 2 for above problem.

###### size - The shape of the returned array.

```python
from numpy import random

x = random.poisson(lam=1, size =10)
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(random.poisson(lam=2, size=1000), kde=False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore", message="The figure layout has changed to tight")

normal_data = random.normal(loc=50, scale=7, size=1000)
sns.kdeplot(normal_data, label='normal')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
sns.kdeplot(random.poisson(lam=50, size=1000), label='poisson')

# Set labels and legend
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()

# Show the plots
plt.show()
```

#### Uniform Distribution
##### Used to describe probability where every event has equal chances of occuring
E.g. Generation of random numbers.

It has three parameters:

a - lower bound - default 0 .0.

b - upper bound - default 1.0.

size - The shape of the returned array.

```python
from numpy import random

uniform_data = random.uniform(size=(10))

print(uniform_data)
```

#### Logistic Distribution
##### Logistic Distribution is used to describe growth.
##### Used extensively in machine learning in logistic regression, neural networks etc.
##### It has three parameters:
##### loc - mean, where the peak is. Default 0.
##### scale - standard deviation, the flatness of distribution. Default 1.
##### size - The shape of the returned array. ##### 

```python
from numpy import random

x = random.logistic(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.kdeplot(random.logistic(size=1000),label="Logistic")
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# difference between logistic and normal distribution

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.kdeplot(random.normal(scale=2, size=1000), label='normal')
sns.kdeplot(random.logistic(size=1000), label='logistic')
plt.legend()
plt.show()
```

###### Write a NumPy program to test whether none of the elements of a given array are zero

```python
import numpy as np
from numpy import random

arr1 = random.randint(100,size=(10))
print(arr1)
x = np.where(arr1==0)
print(x)
```

###### Write a NumPy program to test if any of the elements of a given array are non-zero. 

```python
import numpy as np
from numpy import random

arr = random.randint(2,size=(10))
non_zero = np.where(arr!=0)
print(arr)
print(non_zero)
```
