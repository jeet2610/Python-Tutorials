# Linear Regression with Examples

```python
import matplotlib.pyplot as plt
from scipy import stats
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
slope, intercept, r,p,std_err = stats.linregress(x,y)
def myfunc(x):
    return slope * x + intercept
mymodel = list(map(myfunc,x))

plt.xlabel("Age of the car")
plt.ylabel("Speed of the Car")
plt.scatter(x,y)
plt.scatter(x,mymodel,marker="o",color='red')
plt.show()
```

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
df = pd.read_csv('dataset/salary_data.csv')
```

```python
# salary with avg value
avg_value = df['Salary'].mean()
df['Salary'].fillna(avg_value,inplace=True)
# experience with avg value
avg_experience = df['Years of Experience'].mean()
df['Years of Experience'].fillna(avg_experience,inplace=True)
```

```python
salary = df['Salary']
years= df['Years of Experience']
```

```python
df.info()
```

```python
slope, intercept, r ,p , std_err = stats.linregress(years,salary)
```

```python
print("Relation of Coefficient: ",r)
print("Std_err: ",std_err)
```

```python
def myfunc(years):
    return slope * years + intercept
mymodel = list(map(myfunc,years))
```

```python
plt.scatter(years,salary,marker="s",color='green',label="actual")
plt.plot(years,mymodel,marker="o",color="blue",label="predicted")
plt.xlabel(" Salary")
plt.ylabel("Years of Experience")
plt.title("Years of Experience vs Salary")
plt.legend()
plt.show()
```
