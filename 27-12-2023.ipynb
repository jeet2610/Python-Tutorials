```python
# Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

x1 = np.random.randint(20,size=10)
y1 = np.random.randint(100,size=10)
plt.xlim(0,20)
plt.ylim(0,101)
plt.scatter(x1,y1,color="red")
plt.show()
```

```python
# Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes
import random
import matplotlib.pyplot as plt
import math

no_of_balls = 100
x = [random.triangular() for i in range(no_of_balls)]
y = [random.gauss(0.5,0.25) for i in range(no_of_balls)]
colors = [random.randint(1,4) for i in range(no_of_balls)]
areas = [math.pi * random.randint(3,15)**2 for i in range(no_of_balls)]

plt.figure()
plt.scatter(x,y,s=areas,c=colors,alpha=0.85)
plt.show()
```

# Stem Gallery

```python
#adding a stem chart
import matplotlib.pyplot as plt
import numpy as np
custom_width = 8  # in inches
custom_height = 6  # in inches

# Create a figure with custom width and height
plt.figure(figsize=(custom_width, custom_height))


# make data
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]
#plot
fig,ax = plt.subplots()
ax.stem(x,y,linefmt="-.",orientation='horizontal',bottom=0.5,label="Data Points")
ax.set(xlim=(0,8),xticks=np.arange(0,9),
       ylim=(0,8),yticks=np.arange(1,8))
plt.show()
```

```python
# plots vertical lines from a baseline to the y-coordinate and places a marker at the tip.
import matplotlib.pyplot as plt
import numpy as np
custom_width = 8  # in inches
custom_height = 6  # in inches

# Create a figure with custom width and height
plt.figure(figsize=(custom_width, custom_height))

x = np.linspace(0.1,np.pi*2,41)
y = np.exp(np.sin(x))
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
custom_width = 8  # in inches
custom_height = 6  # in inches

# Create a figure with custom width and height
plt.figure(figsize=(custom_width, custom_height))

x = np.linspace(0.1,np.pi*2,41)
y = np.exp(np.sin(x))
markerline= plt.stem(x,y,linefmt='grey',markerfmt='D',bottom=1.1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(14)
y = np.sin(x / 2)
custom_width = 8  # in inches
custom_height = 6  # in inches

# Create a figure with custom width and height
plt.figure(figsize=(custom_width, custom_height))


plt.step(x, y + 2, label='pre (default)')
plt.plot(x, y + 2, 'o--', color='grey', alpha=0.3)

plt.step(x, y + 1, where='mid', label='mid')
plt.plot(x, y + 1, 'o--', color='grey', alpha=0.3)

plt.step(x, y, where='post', label='post')
plt.plot(x, y, 'o--', color='grey', alpha=0.3)

plt.grid(axis='x', color='0.95')
plt.legend(title='Parameter where:')
plt.title('plt.step(where=...)')
plt.show()
```

##### Calculate total sale data for last year for each product and show it using a Pie chart

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('company_sales_data.csv')
month = df['month_number']
sizes = [10,1,100,12,11,15]
colormap = plt.cm.tab20

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesdata = [df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.figure(figsize=(5,4))
plt.pie(salesdata, labels=labels, autopct='%1.1f%%', colors=colormap(sizes))
plt.axis('equal')
plt.legend(loc='lower left')
plt.show()
```
