# Scipy Interpolation


```python
# 1D interpolation is used to interpolate a distribution with 1 variable

from scipy.interpolate import interp1d
import numpy as np
```

```python
xs = np.arange(10)
ys = 2*xs + 1
interp_func = interp1d(xs,ys)
newarr = interp_func(np.arange(2.1,3,0.1))
print(newarr)
```

```python
# Spline Interpolation
from scipy.interpolate import UnivariateSpline
import numpy as np

xs = np.arange(10)
ys = xs**2 + np.sin(xs) + 1

interp_func = UnivariateSpline(xs, ys)

newarr = interp_func(np.arange(2.1, 3, 0.1))

print(newarr)
```

```python
# Interpolation with Radial Basis Function
from scipy.interpolate import Rbf
import numpy as np

xs = np.arange(10)
ys = xs**2 + np.sin(xs) + 1

interp_func = Rbf(xs, ys)

newarr = interp_func(np.arange(2.1, 3, 0.1))

print(newarr)
```

# Significance Test in Scipy

### T-Test

```python
import numpy as np
from scipy.stats import ttest_ind

v1 = np.random.normal(size=100)
v2 = np.random.normal(size=100)

res = ttest_ind(v1,v2)
print(res)
```

```python
res = ttest_ind(v1,v2).pvalue
print(res)
```

```python
res =ttest_ind(v1,v2).statistic
print(res)
```

### KS-Test

```python
# ks is used to check if the given values follow distribution

import numpy as np
from scipy.stats import kstest

v = np.random.normal(size=100)
res = kstest(v,'norm')
print(res)
```

# Statistical Description of Data
It returns the following description:

number of observations (nobs)
<br> minimum and maximum values = minmax
<br>mean
<br>variance
<br>skewness
<br>kurtosis

```python
# Show statistical description of the values in an array:

import numpy as np
from scipy.stats import describe

v = np.random.normal(size=100)
res = describe(v)

print(res)
```

### Normality tests are based on the skewness and kurtosis

```python
import numpy as np
from scipy.stats import skew, kurtosis

v = np.random.normal(size=100)

print(skew(v))
print(kurtosis(v))
```

```python
# Find if the data comes from a normal distribution:

import numpy as np
from scipy.stats import normaltest

v = np.random.normal(size=100)

print(normaltest(v))
```

Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
<br>Test Data:
<br>test.txt
<br>1 2
<br>2 4
<br>3 1

```python
import matplotlib.pyplot as plt
with open("text.txt") as f:
    data = f.read()
data = data.split('\n')
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]

plt.plot(x,y,marker="o")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sample Graph')
plt.show()
```

Write a Python program to plot two or more lines and set the line markers.

```python
import matplotlib.pyplot
# x axis values
x = [1,4,5,6,15]
# y axis values
y = [2,6,3,6,5]

plt.plot(x,y,color="red",linestyle="dashdot",linewidth=3,marker="o",markerfacecolor="blue",markersize=12)
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
#Set the y-limits of the current axes.
plt.ylim(-1,20)
#Set the x-limits of the current axes.
plt.xlim(-1,20)
# giving a title to my graph
plt.title('Display marker')
# function to show the plot
plt.show()
```
