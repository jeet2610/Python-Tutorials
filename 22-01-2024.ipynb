```python
# Datetime Series Example and Test

import pandas as pd
import numpy as np
pdt = pd.date_range(start="2024-01-22",end="2029-01-22",freq="D")
# print(len(pdt))
v = np.random.random(size=len(pdt))
df = pd.DataFrame({"value": v},index=pdt)
print(df)
```

```python
mn = df.resample('M').mean() #Mean by Month
yn = df.resample("Y").mean() #Mean by Year
print(mn)
print(yn)
```

```python
# reset the index after drpping few columns
import pandas as pd
import numpy as np
d = {'col1': [1, 4, 3, 4, 5], 'col2': [4, 5, 6, 7, 8], 'col3': [7, 8, 9, 0, 1]}
df = pd.DataFrame(d)
print("Original DataFrame")
print(df)
df = df.drop(df.index[[1,3]])
print("Before Indexing")
print(df)
df = df.reset_index()
print("After Indexing")
print(df)
```

```python
# Write a Pandas program to divide a DataFrame in a given ratio.
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(20, 2))
print("Original DataFrame:")
print(df)
part_70 = df.sample(frac=0.7,random_state=10)
part_30 = df.drop(part_70.index)
print(part_70)
print("30% of the said DataFrame")
print(part_30)
```

```python
# 39. Write a Pandas program to combining two series into a DataFrame.
import pandas as pd
import numpy as np

s1 = pd.Series(['100', '200', 'python', '300.12', '400'])
s2 = pd.Series(['10', '20', 'php', '30.12', '40'])

print("Original DataFrame")
print(s1,s2)
print("Combined into one ")
df = pd.concat([s1,s2],axis=0)
print(df)
```
