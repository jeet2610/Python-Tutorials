##### Question 1 : Visualize the monthly sales data to illustrate the trend of sales over the months

```python
df = pd.read_csv('dataset/sales_data.csv')
```

```python
months = df['Month']
sales = df['Sales']
#convert month names to categorial data for proper ordering
months_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
df['Month'] = pd.Categorical(df['Month'],categories=months_order,ordered=True)
df = df.sort_values('Month')
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Month'],df['Sales'],marker='o',linestyle='-',color='b')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
#  Adding labels around the marker
for i,sales in enumerate(df['Sales']):
    plt.text(df['Month'][i],sales,str(sales),ha='right',va='bottom',fontsize=8)
    
plt.xticks(rotation=40)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Question 2 : from the dataset expenses.csv.
##### Create a pie chart to visualize the distribution of expenses across different categories.



```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('dataset/expenses.csv')
```

```python
category = data['Category']
amount = data['Amount']
expense_by_category = data.groupby('Category')['Amount'].sum().reset_index()

plt.pie(expense_by_category['Amount'],labels=expense_by_category['Category'],autopct='%1.0f%%',startangle=14)

plt.axis()
plt.tight_layout()
plt.show()
```

#### Question 3 Create a Bar Chart to display the average ratings of different movie genre



```python
import pandas as pd
import matplotlib.pyplot as plt

movie_data = pd.read_csv('dataset/movies_rating.csv')
```

```python
avg_ratings_by_genre = movie_data.groupby('Genre')['Rating'].mean().reset_index()

#Plotting a bar chart for avg ratings by Genre

plt.figure(figsize=(10,6))
plt.bar(avg_ratings_by_genre['Genre'],avg_ratings_by_genre['Rating'],color='skyblue')
plt.title('Average Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.ylim(0,10)
plt.xticks(rotation = 45,ha = 'center')
plt.tight_layout()
plt.show()
```

#### Question 4 : Create a scatter plot to visualize the relationship between the average ratings and genres of movies.

```python
plt.scatter(movie_data['Genre'],movie_data['Rating'],color='green',alpha=0.5,marker="s",s=100)
plt.title("Relationship Between Genre and Rating")
plt.xlabel("Genre")
plt.ylabel("Rating")
plt.xticks(rotation=45,ha='right')
plt.legend()
plt.show()
```

```python
import pandas as pd
import random

# List of book genres
genres = ["Classic", "Fantasy", "Dystopian", "Romance", "Mystery", "Science Fiction", "Thriller", "Historical Fiction"]

# Generate synthetic data for 100 books with random ratings and genres
num_books = 100

books_data = {
    'Book': [],
    'Genre': [],
    'Rating': []
}

for i in range(num_books):
    book_name = f"Book {i+1}"
    genre = random.choice(genres)
    rating = round(random.uniform(3.0, 5.0), 1)
    
    books_data['Book'].append(book_name)
    books_data['Genre'].append(genre)
    books_data['Rating'].append(rating)

# Create a DataFrame and save as CSV
df = pd.DataFrame(books_data)
df.to_csv('dataset/books_ratings_100.csv', index=False)
```

```python
data = pd.read_csv('dataset/books_ratings_100.csv')
book_genre = data['Genre']
book_rating = data['Rating']

plt.bar(book_genre,book_rating,color='orangered')
plt.xticks(rotation=45,ha='right')
plt.show()
```

##### Question 5 :- Create a scatter plot to visualize the relationship between student's age and their score. Customize the markers by assigning different colors to distinguish scores aobve and below 85.


```python
import pandas as pd
import matplotlib.pyplot as plt

d = pd.read_csv('dataset/student_scores.csv')
above_85 = d[d['Score']>85]
below_85 = d[d['Score']<=85]
plt.figure(figsize=(10,6))
plt.scatter(above_85['Age'],above_85['Score'],color='green',label="Score Above 85")
plt.scatter(below_85['Age'],below_85['Score'],color='orangered',label="Score Below 85",edgecolors="black")
plt.title("Students Age vs Test Scores")
plt.xlabel("Age")
plt.ylabel("Test Score")
plt.legend()
plt.tight_layout()
plt.xlim(12.5,15)
plt.ylim(70,100)
plt.show()
# print(below_85)
```

#### Create a grouped bar chart to visualize the average ratings of different cuisine types across multiple restaurants. Additionally, overlay this chart with error bars representing the standard deviation of ratings within each cuisine type.

```python

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('dataset/restaurant_reviews.csv')
```

```python
grouped_data = data.groupby('Cuisine')['Rating'].agg(['mean','std'])

plt.figure(figsize=(10,6))
grouped_data['mean'].plot(kind='bar',yerr=grouped_data['std'],capsize=5,color='grey')
plt.title('Average Ratings by Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.xticks(rotation=45,ha='right')

plt.tight_layout()
plt.show()
          
```

#### Create a scatter plot to analyze the relationship between Service Quality and Response Time based on customer feedback. Additionally, color the data points based on the Overall Satisfaction rating provided by the customers.

```python
#program to genreate customer feedback
import pandas as pd
import numpy as np

# Generating synthetic data for customer feedback
np.random.seed(42)  # For reproducibility

num_customers = 100
customer_ids = range(1, num_customers + 1)
service_quality = np.random.uniform(3.0, 5.0, num_customers)
response_time = np.random.uniform(3.0, 5.0, num_customers)
overall_satisfaction = np.random.uniform(3.0, 5.0, num_customers)

# Creating a DataFrame
data = pd.DataFrame({
    'Customer ID': customer_ids,
    'Service Quality': service_quality,
    'Response Time': response_time,
    'Overall Satisfaction': overall_satisfaction
})

# Saving the synthetic data to a CSV file
data.to_csv('dataset/customer_feedback.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('dataset/customer_feedback.csv')
#service quality and response time by rating


plt.figure(figsize=(10,6))
scatter = plt.scatter(df['Service Quality'],df['Response Time'],c=data['Overall Satisfaction'],cmap='coolwarm',alpha=0.8)
plt.colorbar(scatter,label="Overall Satisfaction")
plt.title("The relationship b/w Service Quality and Response Time")
plt.xlabel("Service Quality")
plt.ylabel("Response Time")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Create a grouped bar chart to compare the average number of tasks completed and hours worked by employees in different departments.

```python
import pandas as pd
import numpy as np

# Generating synthetic data for employee productivity
np.random.seed(42)  # For reproducibility

num_employees = 200
employee_ids = range(1, num_employees + 1)
departments = np.random.choice(['Sales', 'Marketing', 'IT', 'HR', 'Finance'], size=num_employees)
tasks_completed = np.random.randint(10, 50, num_employees)
hours_worked = np.random.randint(20, 50, num_employees)

# Creating a DataFrame
data = pd.DataFrame({
    'Employee ID': employee_ids,
    'Department': departments,
    'Tasks Completed': tasks_completed,
    'Hours Worked': hours_worked
})

# Saving the synthetic data to a CSV file
data.to_csv('dataset/employee_productivity.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dataset/employee_productivity.csv')
df_department = df.groupby('Department').agg({'Tasks Completed':'mean','Hours Worked':'mean'})
```

```python
fig, ax = plt.subplots(figsize=(10,6))
df_department.plot(kind='bar',ax=ax)
plt.title("Average Tasks Completed and Hours Worked by Departments")
plt.xlabel("Departments")
plt.ylabel("Average Count")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True,linestyle=":")
plt.tight_layout()
plt.show()
```
