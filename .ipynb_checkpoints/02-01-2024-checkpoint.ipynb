### subploting

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
x=[1,2,3,4]
y=[1,2,3,4]
axes[0].plot(x,y,color="blue",ls=":",lw=3)
axes[1].plot(x,y,color="red",ls="-",lw=3)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))
x=[1,2,3,4]
y=[1,2,3,4]
axes[0].plot(x,y,color="blue",ls=":",lw=3)
axes[1].plot(x,y,color="red",ls="-",lw=3)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))
x=[1,2,3,4]
y=[1,2,3,4]
axes[0].plot(x,y,color="blue",ls=":",lw=3)
axes[0].set_xlabel("X axis")
axes[0].set_ylabel("Y axis")
axes[1].plot(x,y,color="red",ls="-",lw=3)
axes[1].set_xlabel("X-Axis")
axes[1].set_ylabel("Y-Axis")
plt.show()
```

```python
# four in a subplot
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
x=[1,2,3,4]
y=[1,2,3,4]
axes[0][0].plot(x,y,color="b",ls=':',lw=4)
axes[0][0].set_xlabel("X-axis")
axes[0][0].set_ylabel("Y-axis")
axes[0][0].set_title("Subplot - 1")
axes[0][1].plot(x,y,color="r",ls='-',lw=5)
axes[0][1].set_xlabel("X-axis")
axes[0][1].set_ylabel("Y-axis")
axes[0][1].set_title("Subplot - 2")
axes[1][0].plot(x,y,color="g",ls='-',lw=8)
axes[1][0].set_xlabel("X-axis")
axes[1][0].set_ylabel("Y-axis")
axes[1][0].set_title("Subplot - 3")
axes[1][1].plot(x,y,color='cyan',ls=':',lw=10)
axes[1][1].set_xlabel("X-axis")
axes[1][1].set_ylabel("Y-axis")
axes[1][1].set_title("Subplot - 4")
fig.suptitle("4 Subplots with four diffeerent Styles",size=20)
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axes = plt.subplots(1,2,figsize=(10,4))
axes[0].plot(x,y1,color='skyblue')
axes[0].set_title('Sin Wave')
axes[1].plot(x,y2,color='orangered')
axes[1].set_title('Cosine Wave')
plt.tight_layout()
plt.show()
```

##### read the data from the csv file into a pandas dataframe

```python
#create a line plot to visualize the sales trend for both product A and Product B over the months
```

```python
df = pd.read_csv('dataset/sales_data (2).csv')
plt.plot(df['Month'],df['Product_A_Sales'],color='r',ls='-',lw=2,label="Product_A_Sales",marker="o")
plt.plot(df['Month'],df['Product_B_Sales'],color='g',ls='-',lw=2,label="Product_B_Sales",marker="s")

for i,val in enumerate(df['Product_A_Sales']):
    plt.text(df['Month'][i],val,str(val),ha='left',va='bottom',fontsize=8,color='white',bbox=dict(facecolor='red',alpha=0.9))
for i, val in enumerate(df['Product_B_Sales']):
    plt.text(df['Month'][i],val,str(val),ha='left',va='bottom',fontsize=8,color='white',bbox=dict(facecolor='green',alpha=0.9))
plt.legend()
plt.xlabel("Month")
plt.ylabel('Product Sales')
plt.ylim(100,200)
plt.grid(True,ls=":",lw=1,which="major",color='black')
plt.title('Sales chart for Products')

plt.tight_layout()
plt.show()
```

```python
# Create a stacked bar plot to visualize the dsitribution of monthly expenses for each category
```

```python
df = pd.read_csv('dataset/expense_data.csv')
categories = ['Groceries','Utilities','Transportation','Entertainment']

#create a stack barchart
plt.figure(figsize=(10,6))
bottom=0
custom_color=['lightyellow','skyblue','lightgreen','pink']
for category,color in zip(categories,custom_color):
    
    plt.bar(df['Month'],df[category],label=category,bottom=bottom,color=color)
    bottom = bottom + df[category]

plt.legend(fancybox=True,loc="upper right")
plt.grid(True,ls=":",alpha=0.2,color='black')
plt.show()
```

#### Stacked Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Q1': [200, 300, 150, 250],
    'Q2': [220, 320, 170, 270],
    'Q3': [240, 310, 160, 280],
    'Q4': [230, 330, 180, 260]
}

df = pd.DataFrame(data)

#Create a stackbar chart
colors=['skyblue','lightyellow','lightgreen','grey']
quarters = ['Q1','Q2','Q3','Q4']
bottom=0
plt.figure(figsize=(8,6))

for quarter,colors in zip(quarters,colors):
    plt.bar(df['Region'],df[quarter],label=quarter,bottom=bottom,color=colors)
    bottom = bottom + df[quarter]

plt.legend()
plt.xlabel("Revenue")
plt.ylabel("Regions")
plt.title("Revenue of Regions by Quarter")
plt.grid(True,ls=":",alpha=0.3,color="black")
plt.show()
```

#### Population Distribution by Age group

```python
import pandas as pd
import matplotlib.pyplot as plt
# Sample data
data = {
    'Age Group': ['0-18', '19-35', '36-50', '51+'],
    'Male': [250, 300, 200, 150],
    'Female': [240, 280, 190, 160],
    'Other': [10, 15, 5, 3]
}
#creating stackbar plot

df = pd.DataFrame(data)
genders = ['Male','Female','Other']
bottom = 0
for gender in genders:
    plt.bar(df['Age Group'],df[gender],label=gender,bottom=bottom)
    bottom = bottom + df[gender]
plt.legend()
plt.title("Population Distribution by Age Group")
plt.ylabel("Population")
plt.xlabel("Age Group")
plt.grid(True,alpha=0.3,color='grey')
plt.show()
```

#### Create a stacked bar chart to visualize the sales distribution for each product category over the months.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('dataset/sales_data (3).csv')

products = ['Toys','Clothing','Books','Electronics']
bottom = 0

for product in products:
    plt.bar(df['Month'],df[product],label=product,bottom=bottom)
    bottom = bottom + df[product]
plt.legend()
plt.title("The Sales Distribution for Each Product Category")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True,alpha=0.3,color='grey')
plt.show()
```
