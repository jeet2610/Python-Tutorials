```python
import pandas as pd
```

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup

url = "https://www.bigbasket.com/cl/beverages/?nc=nb"

# Initialize the webdriver
driver = webdriver.Chrome()  # You need to have chromedriver installed and in your PATH

# Open the URL
driver.get(url)

# Scroll to the bottom of the page to load more items
scroll_pause_time = 10 # Adjust this value as needed
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Once all items are loaded, extract the HTML content
html_content = driver.page_source

# Close the webdriver
driver.quit()

# Use BeautifulSoup to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all items using a refined selector
results = soup.find_all('div', {'class': 'SKUDeck___StyledDiv-sc-1e5d9gk-0 eA-dmzP'})

# Now you can proceed with processing the results
for result in results:
    # Process each item as needed
    product_name = result.find('h3', {'class': 'block m-0 line-clamp-2 font-regular text-base leading-sm text-darkOnyx-800 pt-0.5 h-full'})
    product_price = result.find('span', {'class': 'Label-sc-15v1nk5-0 Pricing___StyledLabel-sc-pldi2d-1 gJxZPQ AypOi'})
    product_originalprice = result.find('span', {'class': 'Label-sc-15v1nk5-0 Pricing___StyledLabel2-sc-pldi2d-2 gJxZPQ hsCgvu'})

    print("Product Name:", product_name)
    print("Price:", product_price)
    print("Orginal Price:", product_originalprice)
    print()  # Print an empty line for better readability
```

```python
len(results)
```

```python
product_name = []
product_price = []
product_originalprice = []
for i in range(len(results)):
    product_originalprice_element = results[i].find('span', {'class': 'Label-sc-15v1nk5-0 Pricing___StyledLabel2-sc-pldi2d-2 gJxZPQ hsCgvu'})
    product_name.append(results[i].find('h3', {'class': 'block m-0 line-clamp-2 font-regular text-base leading-sm text-darkOnyx-800 pt-0.5 h-full'}).get_text().strip())
    product_price_element = results[i].find('span', {'class': 'Label-sc-15v1nk5-0 Pricing___StyledLabel-sc-pldi2d-1 gJxZPQ AypOi'})
    if product_price_element is not None:
        product_price.append(product_price_element.get_text().strip())
    else:
        product_price.append("₹0")
    if product_originalprice_element is not None:
        originalprice = product_originalprice_element.get_text().strip()
    else:
        originalprice = "₹0"
    if originalprice == '₹0':
        product_originalprice = product_price
    
    product_originalprice.append(product_originalprice)
#pricing error , from original price 0 to offer price
```

```python
product_originalprice
```

```python
data = {'Product Name':product_name,'Offer Price':product_price,'Original Price':product_originalprice}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Offer Price'].min()
```

```python
df['Offer Price'] = df['Offer Price'].str.replace('None','₹0')
```

```python
df['Original Price'] = df['Original Price'].str.replace('None','₹0')
```

```python
# Price C
df['Offer Price'] = df['Offer Price'].str[1:]
```

```python
df['Offer Price'] = df['Offer Price'].str.replace(',', '').astype(float)
```

```python
# Price Original Price
df['Original Price'] = df['Original Price'].str[1:]
```

```python
df['Original Price'] = df['Original Price'].str.replace(',','').astype(float)
```

```python
df['Discount'] = ((df['Original Price'] - df['Offer Price'])/df['Original Price'])*100
df['Discount']
```

```python
df['Discount'] = df['Discount'].astype(str)
```

```python
df['Discount'] = df['Discount'].str.replace('-inf','0').astype(float)
```
