```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

nifty50 = pd.read_csv('dataset/nifty50.csv')
open_nifty50 =['Open']
reliance = pd.read_csv('dataset/RELIANCE.NS.csv')
```

```python
# reliance

import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

a = reliance['Open']
x = nifty50['Open']
# print(x)
slope, intercept,r, p, std_err = stats.linregress(x,a)
print(r)
def myfunc(x):
    return slope*x+ intercept

mymodel = list(map(myfunc,x))
plt.scatter(x,a)
plt.plot(x,mymodel)
plt.plot()
```
